# Read in the surveys

```python
all_survey = pd.read_csv("tables/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("tables/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pd.concat([all_survey, d75_survey], axis=0)

survey=survey.rename(columns={'dbn':'DBN'})
# DBN or District Borough Number is the combination of the District Number, 
# the letter code for the borough, and the number of the school
survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
```

# Adding DBN columns

```python
data['hs_directory']['DBN'] = data['hs_directory']['dbn']

# concating a 0 at the start of single digit CS{1 becomes 01, 2 becomes 02, etc.}
def pad_csd(num):
    string_representation = str(num)
    if len(string_representation) > 1:
        return string_representation
    else:
        return '0' + string_representation
    
data['class_size']['padded_csd'] = data['class_size']['CSD'].apply(pad_csd)
# Creating a DBN column using the padded csd and school code, see explanation of DBN above
data['class_size']['DBN'] = data['class_size']['padded_csd'] + data['class_size']['SCHOOL CODE']
```

# Converting SAT & AP score columns to numeric

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")
# Calculating total SAT Score
data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data["ap_2010"][col] = pd.to_numeric(data["ap_2010"][col], errors="coerce")

```

# Extracting latitude & longitude

```python
def find_lat(loc):
    coords = re.findall('\(.+, .+\)', loc)
    lat = coords[0].split(',')[0].replace('(', '')
    return lat

def find_lon(loc):
    coords = re.findall('\(.+, .+\)', loc)
    lon = coords[0].split(',')[1].replace(')', '').strip()
    return lon

data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(find_lat)
data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(find_lon)

data['hs_directory']['lat'] = pd.to_numeric(data['hs_directory']['lat'], errors='coerce')
data['hs_directory']['lon'] = pd.to_numeric(data['hs_directory']['lon'], errors='coerce')
```

# Condensing and combining the datasets

```python
class_size = data['class_size']
class_size = class_size.rename(columns={'GRADE ':'GRADE'})

# Only selecting high school students that belong to the 'GEN ED' program type
class_size = class_size[class_size['GRADE'] == '09-12']
class_size = class_size[class_size['PROGRAM TYPE'] == 'GEN ED']

# Grouping by DBN and aggregating by the mean of possible columns
class_size = class_size.groupby('DBN').agg(numpy.mean)
class_size.reset_index(inplace=True)
data['class_size'] = class_size

# Getting the most recent and relevant data from the demographics and graduation tables
data['demographics'] = data['demographics'][data['demographics']['schoolyear'] == 20112012]
data['graduation'] = data['graduation'][data['graduation']['Cohort'] == '2006']
data['graduation'] = data['graduation'][data['graduation']['Demographic'] == 'Total Cohort']

# Beginning the combination of the data by initating with our  table of interest(SAT).
combined = data["sat_results"]

# left joining these datasets in order not to lose any of the SAT table's data
combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

# Inner joining the other datasets which have more relevant features,
# if any of these features are missing for a high school they will not be included.
to_merge = ["class_size", "demographics", "survey", "hs_directory"]
for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")

# Missing numeric values are filled with their mean and non numeric values are filled with 0
combined = combined.fillna(combined.mean(numeric_only=True))
combined = combined.fillna(0)
```

# Add a school district column for mapping

```python
# School district code is the first 2 char of the DBN, see above explanation for DBN.
combined["school_dist"] = combined["DBN"].str[:2]
```

# Looking for correlations

```python
correlations = combined.corr()
correlations = correlations["sat_score"].sort_values(ascending=True)

# sat_score index is dropped along with null values
correlations.drop('sat_score', inplace=True)
correlations.dropna(inplace=True)

# Removing all values that do not have a correlation higher than Â±35%
boolArr = np.logical_and(correlations.values <=.35, correlations.values >= -.35 )
correlationsGraph = correlations[~boolArr]


# Correlations are plotted and displayed
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.barh(correlationsGraph.index,correlationsGraph.values)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
plt.show()

print(correlationsGraph.sort_values(ascending = False))
```

# Correlations between Race & SAT Scores

Schools with higher percentages of asian and white students have a positive correlation with higher SAT Scores, the opposite is true for schools with higher percetnages of hispanic and black students.

A weak negative correlation between the percentage of hispanic students and sat scores is seen above. Another key insight we gain from the above graph is that schools with a very high percetnage of hispanic students (80% or more) never have a higher SAT Score than 1400.
